```python
def npnan_outliers(dataframe):
    df = dataframe.copy()
    
    for col in df.columns:
        
        q1 = df[col].quantile(.25)
        q3 = df[col].quantile(.75)
        iqr = q3 - q1
        up = round(q3+(1.5*iqr))
        down = round(q1-(1.5*iqr))
        
        df[col] = df[col].apply(lambda x: np.nan if x > up else x)
        df[col] = df[col].apply(lambda x: np.nan if x < down else x)

    return df
```

```python
raw_data = pd.read_csv('diamonds.csv')

# dropping depth and table, they are not worth it.
data = raw_data.drop(['depth', 'table'],axis=1)

# Separating values:
data1 = data.query('x ==0 or y == 0 or z== 0')
data2 = data.query('x !=0 and y != 0 and z != 0')

# Converting nummeric data to log.
data2.iloc[:,[0,4,5,6,7]] = data2.iloc[:,[0,4,5,6,7]].apply(lambda x: np.log10(x))

#change outliers for NaNs
data2.iloc[:,[0,4,5,6,7]] = npnan_outliers(data2.iloc[:,[0,4,5,6,7]])

#dropping NaNs
data2 = data2.dropna()

# getting X and y
X = data2.loc[:,['carat', 'cut', 'color', 'clarity', 'x', 'y', 'z']]
X_treat = pd.get_dummies(X, drop_first=True)
y = data2.loc[:,['price']]

#fitting model
model = LinearRegression()
model.fit(X_treat,y)

#predictinc model
y_pred = model.predict(X_treat)

#adding the predicted values
data2['predicted_price'] = y_pred

# put everything back to normal scale
data2.iloc[:,[0,4,5,6,7,8]] = data2.iloc[:,[0,4,5,6,7,8]].apply(lambda x: round(10**x,2))

# now, the other part, which we cannot apply the log for solving, we just "copy" the mean from mike2
data1['predicted_price'] = data2['predicted_price'].mean()

# now join everything back, and reorder columns
data = pd.concat([data1,data2]).sort_index().loc[:,['carat', 'cut', 'color', 'clarity', 'x', 'y', 'z','price','predicted_price']]
```

```python
y = data.price
y_pred = data.predicted_price

mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_true=y, y_pred=y_pred)

print(f'rmse:{rmse}')
print(f'r2:{r2}')
```

```python
raw_data = pd.read_csv('diamonds.csv')

# dropping depth and table, they are not worth it.
data = raw_data.drop(['depth', 'table'],axis=1)

# Separating values:
data1 = data.query('x ==0 or y == 0 or z== 0')
data2 = data.query('x !=0 and y != 0 and z != 0')

#change outliers for NaN
data2.iloc[:,[0,4,5,6,7]] = npnan_outliers(data2.iloc[:,[0,4,5,6,7]])

#dropping NaNs
data2 = data2.dropna()

# Converting nummeric data to log.
data2.iloc[:,[0,4,5,6,7]] = data2.iloc[:,[0,4,5,6,7]].apply(lambda x: np.log10(x))

data_mean = data2.mean()
data_std = data2.std()

#standardizing
data2.iloc[:,[0,4,5,6,7]] = data2.iloc[:,[0,4,5,6,7]].apply(lambda x: (x - x.mean())/x.std())

# getting X and y
X = data2.loc[:,['carat', 'cut', 'color', 'clarity', 'x', 'y', 'z']]
X_treat = pd.get_dummies(X, drop_first=True)
y = data2.loc[:,['price']]

#fitting model
model = LinearRegression()
model.fit(X_treat,y)

#predictinc model
y_pred = model.predict(X_treat)

#adding the predicted values
data2['predicted_price'] = y_pred

#restandarzing
data2['carat'] = (data2['carat'] * data_std['carat']) + data_mean['carat']
data2['price'] = (data2['price'] * data_std['price']) + data_mean['price']
data2['x'] = (data2['x'] * data_std['x']) + data_mean['x']
data2['y'] = (data2['y'] * data_std['y']) + data_mean['y']
data2['z'] = (data2['z'] * data_std['z']) + data_mean['z']
data2['predicted_price'] = (data2['predicted_price'] * data_std['price']) + data_mean['price']

# put everything back to normal scale
data2.iloc[:,[0,4,5,6,7,8]] = data2.iloc[:,[0,4,5,6,7,8]].apply(lambda x: round(10**x,2))

# now, the other part, which we cannot apply the log for solving, we just "copy" the mean from mike2
data1['predicted_price'] = data2['predicted_price'].mean()

# now join everything back, and reorder columns
data = pd.concat([data1,data2]).sort_index().loc[:,['carat', 'cut', 'color', 'clarity', 'x', 'y', 'z','price','predicted_price']]

# calculating score
y = data.price
y_pred = data.predicted_price

mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_true=y, y_pred=y_pred)

print(f'rmse:{rmse}')
print(f'r2:{r2}')
```

# Check score function

Here we will check the score generated for each step. We will define a function to help simplify the process

```python
def get_scores(data, y_pred,rmse_to_compare=0):
    """This functions receives the y_pred for the prices of the diamond dataframe and prints the scores.
    It also prints comparations. If rmse_to_compare is 0, no comparation will be made.
    """
    
    data['predicted'] = y_pred
    mse = mean_squared_error(data.price, data.predicted)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true=data.price, y_pred=data.predicted)
    print(f'RMSE: {round(rmse,3)}')
    print(f'R2 score: {round(r2,5)}')
    
    if rmse_to_compare != 0:
        comparation = round((rmse-rmse_to_compare)/rmse_to_compare*100,2)
        print(f'change in RMSE: {comparation}')
        
        if comparation < 0:
            print('\ngreat success!!')
            
        else:
            print('\nit worsened!!')
    
    return rmse,r2
```

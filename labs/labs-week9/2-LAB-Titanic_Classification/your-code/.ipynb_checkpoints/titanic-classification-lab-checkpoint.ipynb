## Precision-recall scores

## Calculate the precision and recall scores for the test set. 

_hint: Use the methods from sklearn.metrics_ 

https://scikit-learn.org/stable/modules/generated/sklearn.metrics.precision_score.html
https://scikit-learn.org/stable/modules/generated/sklearn.metrics.recall_score.html

```python
tn, fp, fn, tp = confusion_matrix(y_test,y_pred).ravel()

precision = tp / (fp + tp)
recal = tp / (fn + tp)

print("precision :", precision)
print("recal :",round(recal,4))
```

### Questions:

- What do these precision and recall scores mean? Explain precision and recall with your own words.
- Are these the precision and recall scores for which threshold? Explain what a threshold mean with your own words.

```python
"""
precision is, by def, True Positives / (True Positives + False Positives). In our case, it's
how many of the people we correctly thought survived (TP) divided by 
                                            the total people that we think survived (TP + FP).

Titanic case (survived = 1, dead = 0):
TP: people we thought survived and they really survived.
FP: people we thought survived but were in fact dead.

recall (or, sometimes, sensitivity):
recall is, by def, True Positives / (True Positives + False Negatives). In our case, it's
how many of the people we correctly thought survived (TP) divided by 
                                    the total amount of people that actually survived (TP + FN).
                                    
Titanic case (survived = 1, dead = 0):
TP: people we thought survived and they really survived.
FN: people we thought were dead and but they in fact survived.


Those are the falues for a threshold of 50%. That means if our prediction of survival is above 0.5,
we shall consider that this person survived.

On the graph below, we can see how Precision and Recall change, when each one change.
This change happens by changing the threshold value.
A low value makes the recall bigger but the precision higher.

In the example of the titanic, if you lower the threshold, it's like we consider more people
survived, so we rescue a lot of people, but we also rescue more dead (non-survived) people,
and therefore the precision lowers...

"""
```

```python
from sklearn.metrics import plot_precision_recall_curve

plot_precision_recall_curve(logreg, X_test, y_test)
```

# Fixing the Decision Tree algorithm

As you observed, an overfit occurred when using the default values of the DecisionTreeClassifier. This happens because, by default, the Decision Tree is so complex that it manages to memorize the whole dataset despite the granularity necessary to do that. It just keeps creating splits until each observation is memorized. 

In this sense, by default, the Decision Tree algorithm is **too complex**. We will solve this problem by making the tree a bit less complex. We'll change the parameter called `max_depth` for the DecisionTreeClassifier().

Try to find a good value for `max_depth` that helps getting closer results within `train` and `test` scores.

```python
from tqdm import tqdm

df = pd.DataFrame()

for N in tqdm(np.arange(1,10,1)):

    tree = DecisionTreeClassifier(max_depth=N)
    tree.fit(X_train, y_train)

    # TRAIN SET
    tn, fp, fn, tp = confusion_matrix(y_train,tree.predict(X_train)).ravel()

    a1 = tp / (fp + tp)
    a2 = tp / (fn + tp)

    # TEST SET
    tn, fp, fn, tp = confusion_matrix(y_test,tree.predict(X_test)).ravel()

    a3 = tp / (fp + tp)
    a4 = tp / (fn + tp)
    a5 = (tn + tp)/(tn + fp + fn + tp)


    minidf = pd.DataFrame({'depth':N, 'train_precision':a1, 
                          'train_recall':a2, 'test_precision':a3,
                          'test_recall':a4, 'test_score':a5}, index=[0])
    df = pd.concat([df,minidf])
    
df = df.reset_index(drop=True)
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2,figsize=(12,4))

sns.lineplot(x='depth', y='train_precision', data=df, label = 'train_precision',ax=ax[0])
sns.lineplot(x='depth', y='train_recall', data=df, label = 'train_recall',ax=ax[1])
sns.lineplot(x='depth', y='test_precision', data=df, label = 'test_precision',ax=ax[0])
sns.lineplot(x='depth', y='test_recall', data=df, label = 'test_recall',ax=ax[1])
```

```python
"""
Here we see that the the max_depth that makes train and test's precision and recall closer is 1.
The precision and recall values change with max depth, but the precision and recal from each
other get distanced more as the depth rises.

Again, the more depth, the more de train data is memorized, and thus the distance between scores
for train and test gets bigger.

"""
```

# Finally, use more variables

Try to obtain a good score for your problem. Use more variables if you wish. You can plot the tree to understand the variables.

```python
from sklearn.model_selection import train_test_split

#import full dataset
titanic = pd.read_csv('../../../labs-week8/4-LAB-Titanic_Exploration/data/titanic.csv')

# remove name, cabin, ticket and embarked data
titanic = titanic.drop(['Name','Cabin','Ticket','Embarked'],axis=1)

#drop missing
titanic = titanic.dropna()

#define X and y
X = titanic.drop('Survived',axis=1)
y = titanic.loc[:,'Survived']

#get dummies
X = pd.get_dummies(X,drop_first=True)

#train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from tqdm import tqdm

df = pd.DataFrame()

for N in tqdm(np.arange(1,10,1)):

    tree = DecisionTreeClassifier(max_depth=N)
    tree.fit(X_train, y_train)

    # TRAIN SET
    tn, fp, fn, tp = confusion_matrix(y_train,tree.predict(X_train)).ravel()

    a1 = tp / (fp + tp)
    a2 = tp / (fn + tp)

    # TEST SET
    tn, fp, fn, tp = confusion_matrix(y_test,tree.predict(X_test)).ravel()

    a3 = tp / (fp + tp)
    a4 = tp / (fn + tp)
    a5 = (tn + tp)/(tn + fp + fn + tp)


    minidf = pd.DataFrame({'depth':N, 'train_precision':a1, 
                          'train_recall':a2, 'test_precision':a3,
                          'test_recall':a4, 'test_score':a5}, index=[0])
    df = pd.concat([df,minidf])
    
df = df.reset_index(drop=True)
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2,figsize=(12,4))

sns.lineplot(x='depth', y='train_precision', data=df, label = 'train_precision',ax=ax[0])
sns.lineplot(x='depth', y='train_recall', data=df, label = 'train_recall',ax=ax[1])
sns.lineplot(x='depth', y='test_precision', data=df, label = 'test_precision',ax=ax[0])
sns.lineplot(x='depth', y='test_recall', data=df, label = 'test_recall',ax=ax[1])
```

```python
scores = []
df = pd.DataFrame()


for N in tqdm(np.arange(1,30,1)):

    tree = DecisionTreeClassifier(max_depth=N)
    tree.fit(X_train, y_train)

    # TEST SET
    tn, fp, fn, tp = confusion_matrix(y_test,tree.predict(X_test)).ravel()

    a5 = (tn + tp)/(tn + fp + fn + tp)


    minidf = pd.DataFrame({'depth':N,'test_score':a5}, index=[0])
    df = pd.concat([df,minidf])
    
df = df.reset_index(drop=True)


sns.lineplot(x='depth', y='test_score', data=df, label = 'test_score')
```

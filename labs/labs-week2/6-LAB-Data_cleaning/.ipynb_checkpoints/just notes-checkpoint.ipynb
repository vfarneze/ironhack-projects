```python
import pandas as pd
```

```python
my_dict = {'pais': ['Argentina','Brasil','Croacia'],'gold': [1 ,4 ,6]}
my_dict2 = {'pais': ['Argentina','Brasil','Croacia','Dianamarca'],'silver': [1,5,8,0],'bronze':[1,8,6,57]}

df1 = pd.DataFrame(my_dict)
df2 = pd.DataFrame(my_dict2)
display(df1)
df2
```

```python
pd.merge(df1,df2,how='inner')
```

```python
pd.merge(df2,df1,how='inner')
```

```python
pd.merge(df2,df1,how='outer')
```

```python
my_dict = {'country': ['Argentina','Brasil','Croacia'],'silver': [1,5,8],'bronze':[1,8,6]}
my_dict2 = {'country': ['Argentina','Brasil','Croacia','Dianamarca'],'silver': [3,6,9,18],'bronze':[15,98,46,357]}

df1 = pd.DataFrame(my_dict)
df2 = pd.DataFrame(my_dict2)
```

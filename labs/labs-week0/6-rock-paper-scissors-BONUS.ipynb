```python
# First, we import numpy and set the other previous parameters
import numpy as np
gestures = ['rock','paper','scissors','lizard','spock']
n_rounds = 5
rounds_to_win = (n_rounds - 1)/2 + 1
cpu_score = 0
player_score = 0
```

```python
# Then, we need to add a new the function for the new computer gestures
def randomRPS_special():
    '''This function generates an aleatory gesture for the
        computer: rock, paper or scissor, lizard, spock '''
    computer_gesture = np.random.randint(low=0, high=5)
    
    return computer_gesture
```

```python
# now we add a new function for the player:
def RPSplayer_special():
    '''This function just records the player's gesture '''
    user_input = input('please, select \n 0 for "rock", \n 1 for "paper",  \n 2 for "scissors",  \n 3 for "lizard" or \n 4 for "spock":\n')
    player_gesture = int(user_input)
    
    return player_gesture
```

```python
# now we add a new function for deciding the winner
def winnerRPS_special(computer_gesture,player_gesture):
    '''A function that checks who won a round'''
    #conditions for a tie
    if player_gesture == computer_gesture:
        return 0
    
    #conditions for player win
    elif ((player_gesture == 0 and (computer_gesture == 2 or computer_gesture == 3)) or
          (player_gesture == 1 and (computer_gesture == 0 or computer_gesture == 4)) or
          (player_gesture == 2 and (computer_gesture == 1 or computer_gesture == 3)) or
          (player_gesture == 3 and (computer_gesture == 1 or computer_gesture == 4)) or
          (player_gesture == 4 and (computer_gesture == 0 or computer_gesture == 2))):
        return 2
    
   
    #other scenarios player would lose
    else:
        return 1
```

```python
def scores_special():
    '''This function will store the scores of the players'''
    
    print("\n----------The choice of the computer is: %s!----------" %gestures[computer_gesture])
    print("-------------The choice of the player is: %s!-----------" %gestures[player_gesture]) 
    global cpu_score
    global player_score
    
    if winnerRPS_special(computer_gesture,player_gesture) == 1:
        print('\n\ncpu wins!')
        cpu_score += 1
    elif winnerRPS_special(computer_gesture,player_gesture) == 2:
        print('\n\nPLAYER WINS!')
        player_score += 1
    else:
        print('\n\nits a tie!!')

    return cpu_score, player_score
```

```python
#now we redo the game!
cpu_score = 0
player_score = 0
round = 1
print('_____________STARTING_ROUND_#%s_________' %round)
    
while True:
    # Asking for player's choice 
    player_gesture = RPSplayer_special()    
    
    #generate the random choice of the computer 
    computer_gesture = randomRPS_special()
    
    #show the round results 
    cpu_score, player_score = scores_special()      

    #update the scores 
    print('\n--------------------')
    print('\n| CPU SCORE:',cpu_score)
    print('\n| Player SCORE:',player_score)
    print('\n--------------------')
    
    #etc.
    round += 1
    print('_____________STARTING_ROUND_#%s_________' %round)
    
    if (cpu_score == rounds_to_win or player_score == rounds_to_win):
        print('-------end of match!!!!')
        break
    

if cpu_score == rounds_to_win:
    WINNER = "Skynet"
else:
    WINNER = "the player"
```

```python
#printing the winner!
print('The winner is... %s!!!!' %WINNER)
```
